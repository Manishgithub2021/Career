```python
import pandas as pd
from sklearn import tree
model=tree.DecisionTreeClassifier()
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('salaries.csv')
```

```python
df.head()
```

```python
labelencodercompany=LabelEncoder()
labelencoderjob=LabelEncoder()
labelencoderdegree=LabelEncoder()
```

```python
df['newcompany']=labelencodercompany.fit_transform(df.company)
df['newjob']=labelencoderjob.fit_transform(df.job)
df['newdegree']=labelencoderdegree.fit_transform(df.degree)
```

```python
df
```

```python
df.drop(['company','job','degree'],axis=1,inplace=True)
df
```

```python
target=df[['salary_more_then_100k']]
inputs=df.drop(['salary_more_then_100k'],axis=1)
```

```python
target
```

```python
inputs
```

```python
model.fit(inputs,target)
```

```python
model.score(inputs,target)
```
